### Résultats des tests

Afin de garantir la fiabilité et la robustesse du pipeline de traitement des données ainsi que de l’intégration du modèle de machine learning, plusieurs tests unitaires ont été réalisés à l’aide du framework **pytest**. Ces tests ont permis de vérifier les composants essentiels du projet. Tout d’abord, un test a été effectué pour s’assurer que le jeu de données nettoyé ne contient aucune valeur manquante, ce qui confirme que l’étape de prétraitement des données a été correctement réalisée. Ensuite, la fonction `optimize_memory()` a été testée afin de vérifier qu’elle permet effectivement de réduire l’utilisation de la mémoire tout en conservant l’intégrité et la structure des données. Enfin, un test a été réalisé pour vérifier que le modèle de machine learning sauvegardé peut être chargé correctement à partir du fichier `.pkl`, garantissant ainsi le bon fonctionnement du processus de déploiement du modèle. Les résultats obtenus montrent que l’ensemble des tests a été validé avec succès, ce qui confirme que le pipeline de données et l’intégration du modèle sont fiables et fonctionnels.
