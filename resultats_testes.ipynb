##  Automated Testing Results

Afin de garantir la fiabilité du pipeline de traitement des données et de l’intégration du modèle de machine learning, des **tests automatisés** ont été implémentés à l’aide du framework **pytest**. Ces tests permettent de vérifier automatiquement certains composants critiques du projet et de s’assurer que les fonctions principales fonctionnent correctement.

Les tests réalisés dans ce projet sont les suivants :

* **Vérification de la gestion des valeurs manquantes** afin de s'assurer que le dataset nettoyé ne contient pas de données manquantes.
* **Validation de la fonction `optimize_memory(df)`** qui permet de réduire l'utilisation de la mémoire en optimisant les types de données du dataset.
* **Vérification du chargement du modèle de machine learning** afin de confirmer que le modèle sauvegardé (`.pkl`) peut être chargé correctement pour une utilisation ultérieure.

L'exécution de la commande suivante dans le terminal :

```bash
python -m pytest -v -s
```

Le message `warning` affiché est uniquement lié à l’environnement Jupyter et **n’affecte pas le bon fonctionnement du projet**.

Ainsi, ces tests automatisés confirment que les principales composantes du projet — le prétraitement des données, l’optimisation de la mémoire et le chargement du modèle — fonctionnent correctement.
