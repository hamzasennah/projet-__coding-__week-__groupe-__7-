##  Automated Testing Results

Dans le cadre de la validation du projet, des **tests automatisés** ont été implémentés à l’aide du framework **pytest** afin de vérifier la fiabilité du pipeline de traitement des données et l’intégration du modèle de machine learning. Trois tests principaux ont été réalisés. Le premier test permet de vérifier l’absence de **valeurs manquantes** dans le jeu de données nettoyé, garantissant ainsi la qualité du prétraitement des données. Le second test valide le bon fonctionnement de la fonction `optimize_memory(df)`, qui vise à réduire l’utilisation de la mémoire tout en conservant l’intégrité des données. Enfin, un troisième test confirme que le **modèle de machine learning peut être chargé correctement** à partir du fichier sauvegardé (`.pkl`).

L’exécution des tests montre que **les trois tests ont été validés avec succès** (`3 passed`), ce qui confirme que les principales composantes du projet fonctionnent correctement. ,
