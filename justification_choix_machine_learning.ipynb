## Justification du choix du modèle final

Après avoir entraîné et évalué trois modèles de machine learning — **Random Forest**, **XGBoost** et **LightGBM** — plusieurs métriques de performance ont été utilisées pour comparer leurs résultats : **Accuracy, Precision, Recall, F1-score et ROC-AUC**.

Les trois modèles ont obtenu des performances élevées sur le jeu de données, ce qui montre qu’ils sont tous capables de bien capturer les relations entre les variables et la variable cible. Cependant, une analyse comparative des métriques révèle que **LightGBM** présente la **meilleure performance globale**.

En effet, LightGBM obtient un **score ROC-AUC d’environ 0.999**, ce qui indique une excellente capacité à **distinguer les différentes classes**. De plus, ses scores d’**accuracy**, de **precision**, de **recall** et de **F1-score** sont également très élevés et comparables, voire légèrement supérieurs, à ceux obtenus par **Random Forest** et **XGBoost**.

Au-delà des performances, **LightGBM** est également reconnu pour ses **avantages computationnels**. Ce modèle est particulièrement **rapide à entraîner**, **efficace sur les données structurées** et capable de **gérer de grands volumes de données avec une consommation mémoire optimisée** grâce à son algorithme de **Gradient Boosting basé sur des histogrammes**.

Compte tenu de ces éléments — **excellentes performances prédictives**, **meilleure capacité de discrimination (ROC-AUC)** et **efficacité computationnelle** — **LightGBM a été sélectionné comme modèle final** pour ce projet.

Le modèle entraîné a ensuite été **sauvegardé dans le fichier `best_model.pkl`** afin d’être réutilisé ultérieurement dans l’application d’interface et pour générer les prédictions.

### Conseil

Dans un projet de machine learning, il est important de comparer plusieurs modèles et plusieurs métriques de performance avant de choisir le modèle final. L’utilisation de métriques comme **ROC-AUC**, **F1-score** et **Accuracy** permet d’avoir une vision complète de la performance du modèle et d’éviter de sélectionner un modèle basé sur une seule métrique.
