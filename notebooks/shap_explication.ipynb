# 📊 Interprétation des Graphiques SHAP

---

## 📌 Graphique 1 — Feature Importance (Importance Globale des Variables)

> **Objectif :** Identifier quelles caractéristiques influencent le plus la prédiction du niveau d'obésité.

Le premier graphique représente l'**importance globale des variables** dans le modèle à l'aide des valeurs SHAP.
On observe que :

| Variable | Importance | Signification médicale |
|---|---|---|
| **Weight** (poids) | ⭐⭐⭐⭐⭐ Très élevée | Directement lié à l'IMC |
| **Height** (taille) | ⭐⭐⭐⭐ Élevée | Directement lié à l'IMC |
| **Gender** (genre) | ⭐⭐⭐ Moyenne | Influence la distribution du poids |
| **Age** (âge) | ⭐⭐⭐ Moyenne | Lié aux changements métaboliques |
| Habitudes alimentaires | ⭐⭐ Faible | Influence indirecte |
| Mode de vie | ⭐⭐ Faible | Influence indirecte |

Cela signifie que le modèle utilise principalement ces facteurs pour prédire le niveau d'obésité d'un individu.
Ce résultat est cohérent avec les connaissances médicales, car le **poids** et la **taille** sont directement
liés à l'**indice de masse corporelle (IMC)**, un indicateur clé pour évaluer l'obésité.

Les autres variables, comme les habitudes alimentaires ou le mode de vie, ont une influence plus faible
mais contribuent tout de même à améliorer la précision du modèle.

> 💡 **À retenir :** Plus la barre est longue dans le graphique, plus la variable a d'influence sur la décision du modèle.

## 🔬 Graphique 2 — SHAP Summary Plot (Impact par Individu)

> **Objectif :** Comprendre comment chaque variable influence la prédiction pour chaque individu.

Le second graphique est un **SHAP summary plot** qui montre l'impact de chaque variable sur les prédictions
du modèle pour **chaque individu** du jeu de données.

### 🎨 Comment lire ce graphique ?

- **Chaque point** = une observation (un patient)
- 🔴 **Couleur rouge** = valeur élevée de la variable
- 🔵 **Couleur bleue** = valeur faible de la variable
- **Position à droite** → la variable *augmente* la probabilité d'obésité
- **Position à gauche** → la variable *diminue* la probabilité d'obésité

### 📋 Exemples d'interprétation

| Variable | Valeur élevée | Valeur faible |
|---|---|---|
| **Weight** | 🔴 Augmente fortement le risque | 🔵 Diminue le risque |
| **FCVC** (légumes) | 🔴 Contribue à réduire le risque | 🔵 Augmente légèrement le risque |
| **FAF** (activité physique) | 🔴 Réduit le risque d'obésité | 🔵 Augmente le risque |

Ce graphique permet donc de comprendre non seulement **quelles variables sont importantes**,
mais aussi **comment elles influencent concrètement** les prédictions du modèle.

> 💡 **À retenir :** Ce graphique offre une vision individuelle de l'impact des variables,
> contrairement au graphique 1 qui donne une vision globale.
